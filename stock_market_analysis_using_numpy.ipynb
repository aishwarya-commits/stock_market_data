## You are given a dataset containing the daily closing prices of a stock over a period of  days.

The dataset is stored in a CSV file named  stock_prices.csv , with a single column named  Close . Your task is to perform the following operations using NumPy:

Load the dataset into a NumPy array.

Calculate the daily returns of the stock. Daily returns are calculated as the percentage change from one day to the next

Note: The first day will not have a return value since there is no previous day to compare it with.

Compute the mean and standard deviation of the daily returns.

Identify the top 5 days with the highest positive returns and the top 5 days with the highest negative returns. Provide the indices of these days in the original dataset.

Plot the histogram of the daily returns using Matplotlib. Ensure the histogram has a title, labels for the x-axis and y-axis, and a grid.

Provide the complete code to perform these tasks, including the necessary imports.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import pandas as pd

df = pd.read_csv("stock_data.csv")
print(df)
```

```python
## convert  closing prices into numpy array
import numpy as np

close_prices = df["close"].to_numpy()
print(close_prices)

```

```python
daily_returns = (close_prices[1:] - close_prices[:-1]) / close_prices[:-1] ## calculate daily returns
print(daily_returns)
```

```python
## Step 4: Mean & standard deviation
mean_return = np.mean(daily_returns)
std_return = np.std(daily_returns)

print("Mean Daily Return:", mean_return)
print("Standard Deviation:", std_return)
```

```python
# Top 5 positive and negative returns
top_5_positive = np.argsort(daily_returns)[-5:][::-1]
top_5_negative = np.argsort(daily_returns)[:5]

print("\nTop 5 Positive Return Days:")
for i in top_5_positive:
    print("Day", i + 1, "Return:", daily_returns[i] * 100, "%")

print("\nTop 5 Negative Return Days:")
for i in top_5_negative:
    print("Day", i + 1, "Return:", daily_returns[i] * 100, "%")
```

```python

## Histogram
plt.hist(daily_returns, bins=50)
plt.title("Histogram of Daily Stock Returns")
plt.xlabel("Daily Return")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```
